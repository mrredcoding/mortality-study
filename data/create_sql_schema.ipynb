## Create SQL Schema

```python
def infer_sql_dtype(dtype) -> str:
    """
    Infers the appropriate SQL data type based on a Pandas column data type.
    
    Parameters:
    dtype (dtype): Pandas data type to be converted.
    
    Returns:
    str: Corresponding SQL data type as a string.
    """
    if pd.api.types.is_integer_dtype(dtype):
        return "INT"
    elif pd.api.types.is_float_dtype(dtype):
        return "FLOAT"
    elif pd.api.types.is_bool_dtype(dtype):
        return "BOOLEAN"
    elif pd.api.types.is_datetime64_dtype(dtype):
        return "DATETIME"
    else:
        return "VARCHAR(6)"
    
def generate_sql_schema_from_df(df: pd.DataFrame, table_name: str) -> str:
    """
    Generates an SQL CREATE TABLE statement dynamically from a Pandas DataFrame.
    
    Parameters:
    df (pd.DataFrame): The DataFrame whose structure will be used to define the schema.
    table_name (str): The name of the SQL table to create.
    
    Returns:
    str: SQL CREATE TABLE statement as a string.
    """
    sql_schema = f"CREATE TABLE {table_name} (\n"
    columns_definitions = []
    
    for col in df.columns:
        dtype = infer_sql_dtype(df[col].dtype)
        columns_definitions.append(f"    {col} {dtype}")

    sql_schema += ",\n".join(columns_definitions) + "\n);"
    return sql_schema
```

```python
mortality_rate_name = "MORTALITY_RATE"

import os
base_dir = os.getcwd()
schema_dir = os.path.join(base_dir, "servers", "schema")
```

```python
mortality_rate_schema = generate_sql_schema_from_df(MORTALITY_RATE, mortality_rate_name)
mortality_rate_schema += f"\n\nALTER TABLE {mortality_rate_name} ADD CONSTRAINT mortality_rate_PK PRIMARY KEY (Age, Gender);"

file_path = os.path.join(schema_dir, "mortality_rate_schema.sql")
with open(file_path, "w") as file:
    file.write(mortality_rate_schema)
```

## Insert data

```python
def generate_insert_statements(df: pd.DataFrame, table_name: str) -> str:
    """
    Generates SQL INSERT statements from a Pandas DataFrame.
    
    Parameters:
    df (pd.DataFrame): The DataFrame containing data to insert.
    table_name (str): The name of the SQL table to insert into.
    
    Returns:
    str: SQL INSERT statements as a string.
    """
    insert_sql = ""
    for _, row in df.iterrows():
        values = ', '.join([f"'{x}'" if isinstance(x, str) else str(x) for x in row])
        insert_sql += f"INSERT INTO {table_name} ({', '.join(df.columns)}) VALUES ({values});\n"
    return insert_sql
```

```python
mortality_rate_inserts = generate_insert_statements(MORTALITY_RATE, mortality_rate_name)

file_path = os.path.join(schema_dir, "mortality_rate_inserts.sql")

with open(file_path, "w") as file:
    file.write(mortality_rate_inserts)
```
